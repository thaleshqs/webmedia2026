```python
import pandas as pd
import numpy as np

df = pd.read_csv('classificacao_topicgpt.csv')
df
```

```python
# Certifique-se de que as colunas numéricas estão no formato correto
colunas_metricas = ['view_count', 'like_count', 'comment_count']
for col in colunas_metricas:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# 1. Contagem do número de elementos por categoria (tópico)
contagem_topicos = df.groupby('topico_llm').size().reset_index(name='qtd_elementos')

# 2. Métricas Absolutas: Soma total
metricas_absolutas = df.groupby('topico_llm')[colunas_metricas].sum().reset_index()
metricas_absolutas.rename(columns={
    'view_count': 'total_views',
    'like_count': 'total_likes',
    'comment_count': 'total_comments'
}, inplace=True)

# 3. Métricas Normalizadas (AGORA COM MEDIANA): Mediana por elemento
# A mediana encontra o valor central, ignorando distorções de vídeos super virais ou zerados
metricas_normalizadas = df.groupby('topico_llm')[colunas_metricas].median().reset_index()
metricas_normalizadas.rename(columns={
    'view_count': 'mediana_views_por_elemento',
    'like_count': 'mediana_likes_por_elemento',
    'comment_count': 'mediana_comments_por_elemento'
}, inplace=True)

# 4. Consolidar os resultados em um único DataFrame
df_analise = contagem_topicos.merge(metricas_absolutas, on='topico_llm')
df_analise = df_analise.merge(metricas_normalizadas, on='topico_llm')

# 5. Calcular a Taxa de Engajamento Relativo [(likes + comentários) / views]
# Mantemos o uso dos valores absolutos agregados para mostrar o engajamento geral do tópico
df_analise['engajamento_relativo'] = np.where(
    df_analise['total_views'] > 0,
    (df_analise['total_likes'] + df_analise['total_comments']) / df_analise['total_views'],
    0 # Se tiver 0 views, o engajamento relativo é 0
)

# Coluna Percentual opcional
df_analise['engajamento_relativo_pct'] = (df_analise['engajamento_relativo'] * 100).round(2).astype(str) + '%'

# Ordenar pelo tópico com maior engajamento relativo
df_analise = df_analise.sort_values(by='engajamento_relativo', ascending=False)

# Exibir os resultados
print("Análise de Engajamento por Tópico (com Mediana e Engajamento Relativo):")
display((df_analise)) # Use print(df_analise) se não estiver no Jupyter/Colab
```

```python
df_analise.to_csv('topicgpt_analise.csv')
```
