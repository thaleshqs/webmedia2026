```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd

df_videos = pd.read_json('./videos.jsonl', lines=True).drop_duplicates(subset=['video_id']).reset_index(drop=True)
print(len(df_videos))
df_videos.head()
```

```python
df_comments = pd.read_json('./comments.jsonl', lines=True)
print(len(df_comments))
df_comments.head()
```

```python
# ----------------------------------------------------------------------------
# 0. CONFIGURAÇÃO — ajuste o caminho do arquivo de entrada aqui
# ----------------------------------------------------------------------------
OUTPUT_DIR = Path("output_caracterizacao")
OUTPUT_DIR.mkdir(exist_ok=True)

def prepare_data(df: pd.DataFrame) -> pd.DataFrame:
    """Conversões básicas de tipo e checagens de qualidade dos dados."""
    df = df.copy()

    for col in ["view_count", "like_count", "comment_count"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    date_col = "published_date" if "published_date" in df.columns else "published_at"
    df["published_dt"] = pd.to_datetime(df[date_col], errors="coerce", utc=True)
    df["year"] = df["published_dt"].dt.year

    n_bad_views = df["view_count"].isna().sum()
    n_bad_dates = df["published_dt"].isna().sum()
    if n_bad_views or n_bad_dates:
        print(f"[AVISO] {n_bad_views} linha(s) com view_count inválido; "
              f"{n_bad_dates} linha(s) com data de publicação inválida.")

    return df


# ----------------------------------------------------------------------------
# 1. CDFs com percentis destacados
# ----------------------------------------------------------------------------
def plot_cdf(df: pd.DataFrame, column: str, label: str, filename: str,
             percentiles=(50, 75, 90, 95)):

    data = df[column].dropna()
    data = data[data >= 0]

    x = np.sort(data.values)
    y = np.arange(1, len(x) + 1) / len(x)

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(x, y, color="#1f77b4", linewidth=2)

    colors = ["#d62728", "#2ca02c", "#9467bd", "#363737"]
    for p, c in zip(percentiles, colors):
        val = np.percentile(data, p)
        ax.axvline(val, color=c, linestyle="--", linewidth=1.2)
        ax.axhline(p / 100, color=c, linestyle=":", linewidth=0.8, alpha=0.5)
        ax.annotate(
            f"P{p} = {val:,.0f}",
            xy=(val, p / 100),
            xytext=(8, -12 if p != percentiles[-1] else 3),
            textcoords="offset points",
            color=c,
            fontsize=12,
            fontweight="bold",
        )

    if (x > 0).all():
        ax.set_xscale("log")
        ax.set_xlabel(f"{label} (escala log)", fontsize=20)
    else:
        ax.set_xscale("symlog", linthresh=1.0)
        ax.set_xlabel(f"{label} (escala symlog)", fontsize=18)
        ax.set_xlim(left=0)

    ax.set_ylabel("Probabilidade acumulada (CDF)", fontsize=18)
    #ax.set_title(title)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    ax.set_ylim(0, 1.02)
    ax.grid(True, which="both", linestyle="-", alpha=0.25)
    fig.tight_layout()
    plt.show()
    fig.savefig(OUTPUT_DIR / filename, dpi=400)
    plt.close(fig)
    print(f"[OK] Figura salva em {OUTPUT_DIR / filename}")


# ----------------------------------------------------------------------------
# 2. Volume de vídeos por ano
# ----------------------------------------------------------------------------
def plot_videos_per_year(df: pd.DataFrame, filename: str = "videos_por_ano.png") -> pd.Series:
    counts = df["year"].dropna().astype(int).value_counts().sort_index()

    fig, ax = plt.subplots(figsize=(10, 7))
    bars = ax.bar(counts.index.astype(str), counts.values, color="#1f77b4")
    for bar, val in zip(bars, counts.values):
        ax.annotate(f"{val:,}", xy=(bar.get_x() + bar.get_width() / 2, val),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", fontsize=18)

    ax.set_xlabel("Ano de publicação", fontsize=22)
    ax.set_ylabel("Número de vídeos", fontsize=22)
    #ax.set_title("Volume de vídeos publicados por ano")
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    ax.grid(True, axis="y", linestyle="-", alpha=0.25)
    fig.tight_layout()
    fig.savefig(OUTPUT_DIR / filename, dpi=300)
    plt.close(fig)
    print(f"[OK] Figura salva em {OUTPUT_DIR / filename}")

    counts.to_csv(OUTPUT_DIR / "videos_por_ano.csv", header=["n_videos"])
    return counts


# ----------------------------------------------------------------------------
# MAIN
# ----------------------------------------------------------------------------
def main():
    df = prepare_data(df_videos)

    print(f"Dataset carregado: {len(df):,} vídeos\n")

    print("=" * 70)
    print("1) CDFs de view_count e comment_count")
    print("=" * 70)
    plot_cdf(df, "view_count", "# de visualizações por vídeo","cdf_view_count.png")
    plot_cdf(df, "comment_count", "# de comentários por vídeo","cdf_comment_count.png")
    comments_per_user = (df_comments.groupby("author_channel_id")["comment_id"].count().reset_index(name="comment_count"))
    plot_cdf(comments_per_user, "comment_count", "# de comentários por usuário", "cdf_comment_count_users.png")
    print("\n" + "=" * 70)
    print("2) Volume de vídeos por ano")
    print("=" * 70)
    plot_videos_per_year(df)


if __name__ == "__main__":
    main()
```

```python
def plot_total_views_per_year(df: pd.DataFrame,
                                filename: str = "total_views_por_ano.png") -> pd.Series:
    """Soma o view_count por ano de publicação e plota como gráfico de linhas."""
    totals = df.groupby("year")["view_count"].sum().sort_index()
    totals.index = totals.index.astype(int)
 
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(totals.index.astype(str), totals.values,
            marker="o", color="#1f77b4", linewidth=2)
 
    for x, val in zip(totals.index.astype(str), totals.values):
        ax.annotate(f"{val:,.0f}", xy=(x, val), xytext=(0, 8),
                    textcoords="offset points", ha="center", fontsize=12)
 
    ax.set_xlabel("Ano de publicação", fontsize=18)
    ax.set_ylabel("Visualizações (milhões)", fontsize=18)
    ax.grid(True, axis="y", linestyle="-", alpha=0.25)
    fig.tight_layout()
    fig.savefig(OUTPUT_DIR / filename, dpi=300)
    plt.show()
    plt.close(fig)
    print(f"[OK] Figura salva em {OUTPUT_DIR / filename}")
 
    totals.to_csv(OUTPUT_DIR / "total_views_por_ano.csv", header=["total_views"])
    return totals
df_videos['year'] = df_videos['published_at'].dt.year
plot_total_views_per_year(df_videos)
```

```python
def plot_videos_per_year(df: pd.DataFrame, filename: str = "videos_por_ano.png") -> pd.Series:
    counts = df["year"].dropna().astype(int).value_counts().sort_index()
 
    fig, ax = plt.subplots(figsize=(8, 5))
    bars = ax.bar(counts.index.astype(str), counts.values, color="#1f77b4")
    for bar, val in zip(bars, counts.values):
        ax.annotate(f"{val:,}", xy=(bar.get_x() + bar.get_width() / 2, val),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", fontsize=9)
 
    ax.set_xlabel("Ano de publicação")
    ax.set_ylabel("Número de vídeos")
    ax.set_title("Volume de vídeos publicados por ano")
    ax.grid(True, axis="y", linestyle="-", alpha=0.25)
    fig.tight_layout()
    fig.savefig(OUTPUT_DIR / filename, dpi=200)
    plt.show()
    plt.close(fig)
    print(f"[OK] Figura salva em {OUTPUT_DIR / filename}")
 
    counts.to_csv(OUTPUT_DIR / "videos_por_ano.csv", header=["n_videos"])
    return counts
plot_videos_per_year(df_videos)
```
